```python
# Loading dataset from URL

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"

# Defining column names
column_names = [
    'Pregnancies',
    'Glucose',
    'BloodPressure',
    'SkinThickness',
    'Insulin',
    'BMI',
    'DiabetesPedigreeFunction',
    'Age',
    'Outcome'  # 0 = No Diabetes, 1 = Diabetes
]

print(f"Downloading and loading data from public URL...")
try:
    df = pd.read_csv(url, header=None, names=column_names)
    print("✅ Data successfully loaded!")

except:
    print("Data not Uploaded")
```

```python
# Analylising data

print("\n--- Data Head ---")
print(df.head())

# Basic info of Data
print("\n--- Data Info ---")
df.info()

# statistical summary
print("\n--- Statistical Summary ---")
print(df.describe())
```

```python
# --- Step 3: Data Visualization ---
print("\nGenerating plots...")
sns.set_style("whitegrid")

# 1. Plot:(Distribution)
plt.figure(figsize=(7, 5))
sns.countplot(x='Outcome', data=df)
plt.title('Diabetes Outcome (0 = No, 1 = Yes)')
plt.xlabel('Outcome')
plt.ylabel('Number of Patients')
plt.xticks([0, 1], ['No Diabetes', 'Diabetes'])
plt.show()
```

```python
# 2. Plot: Age distribution (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30, color='purple')
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Plot: Glucose level distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Glucose', hue='Outcome', kde=True, bins=30)
plt.title('Glucose Level Distribution by Outcome')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.legend(title='Outcome', labels=['Diabetes (1)', 'No Diabetes (0)'])
plt.show()
```

```python
# 4. Plot: BMI vs. Age (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='BMI', hue='Outcome', data=df, alpha=0.7)
plt.title('BMI vs. Age (Colored by Diabetes Outcome)')
plt.xlabel('Age')
plt.ylabel('BMI (Body Mass Index)')
plt.legend(title='Outcome', labels=['Diabetes (1)', 'No Diabetes (0)'])
plt.show()

```

```python
# 5. Plot: Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation Heatmap of All Features')
plt.show()
```
